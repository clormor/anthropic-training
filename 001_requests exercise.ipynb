```python
from dotenv import load_dotenv

load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

# Helper function

This function stores state in multi-turn conversations, allowing you to persist and send conversation history.

This is critical, since the Anthropic API is entirely stateless.

```python
def chat(messages, user_message):
    messages.append({"role": "user", "content": user_message})
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    text_block = next((b for b in message.content if b.type == "text"), None)
    if text_block is None:
        raise ValueError(f"No text block in API response: {message.content}")
    response = text_block.text
    messages.append({"role": "assistant", "content": response})
    return response
```

# Chatbot

Create a chatbot using a while True loop, which reads input from the user.

```python
messages = []

while True:
    user_input = input("> ")

    if not user_input.strip():
        continue

    if user_input.strip().lower() == "quit":
        break

    answer = chat(messages, user_input)
    print("---")
    print("> ", answer)
    print("---")
```
