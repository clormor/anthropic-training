```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

# Helper functions

These functions help us to store state in multi-turn conversations, allowing you to persist and send conversation history.

This is critical, since the Anthropic API is entirely stateless.

```python
# Store a message (i.e. a request) from the user
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

# Store a message (i.e. a response) from the API
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

# Send a new request, containing our latest user message appended to the conversation history
def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
```

# Chatbot

Create a chatbot using a while(True) loop, which reads input from the user.

```python
# Start with an empty message list
messages = []

while True:
    user_input = input("> ")
    print("> ", user_input)

    if not user_input.strip():
        continue

    if user_input.strip().lower() == "quit":
        break

    add_user_message(messages, user_input)
    answer = chat(messages)
    print('---')
    print("> ", answer)
    print('---')

    add_assistant_message(messages, answer)
```
