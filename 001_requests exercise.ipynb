```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

# Helper functions

These functions help us to store state in multi-turn conversations, allowing you to persist and send conversation history.

This is critical, since the Anthropic API is entirely stateless.

```python
def add_user_message(messages, text):
    messages.append({"role": "user", "content": text})

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    response = message.content[0].text
    messages.append({"role": "assistant", "content": response})
    return response
```

# Chatbot

Create a chatbot using a while True loop, which reads input from the user.

```python
messages = []

while True:
    user_input = input("> ")
    print("> ", user_input)

    if not user_input.strip():
        continue

    if user_input.strip().lower() == "quit":
        break

    add_user_message(messages, user_input)
    answer = chat(messages)
    print("---")
    print("> ", answer)
    print("---")
```
