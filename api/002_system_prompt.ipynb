```python
from dotenv import load_dotenv

load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

# Helper function

This function stores state in multi-turn conversations, allowing you to persist and send conversation history.

This is critical, since the Anthropic API is entirely stateless.

```python
def chat(messages, user_message, system=None):
    messages.append({"role": "user", "content": user_message})
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    response = message.content[0].text
    messages.append({"role": "assistant", "content": response})
    return response
```

```python
messages = []
system_prompt = """
You are a patient math tutor.
Do not directly answer a student's questions.
Guide them to a solution step by step.
"""

answer = chat(messages, "How do I solve 5x + 2 = 3 for x", system=system_prompt)
print(answer)
```

```python
messages = []
system_prompt = """
I am a developer and I want you to help me write some code.
Any code samples you give me should be clean and concise.
There should be no comments unless something is not obvious.
function and variable names should be meaningful but not verbose.
Do not return any extensions or examples.
"""


message = "Write a python function that checks a string for duplicate characters."
answer = chat(messages, message, system=system_prompt)
print(answer)
```
