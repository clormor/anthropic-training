```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

# Helper functions

These functions help us to store state in multi-turn conversations, allowing you to persist and send conversation history.

This is critical, since the Anthropic API is entirely stateless.

```python
# Store a message (i.e. a request) from the user
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

# Store a message (i.e. a response) from the API
def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

# Send a new request, containing our latest user message appended to the conversation history
def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
```

# Putting it all together

We can use these functions to begin a multi-turn conversation with the Anthropic API that maintains state.

```python
# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "Define quantum computing in one sentence")

# Get Claude's response
answer = chat(messages)

# Add Claude's response to the conversation history
add_assistant_message(messages, answer)

# Add a follow-up question
add_user_message(messages, "Write another sentence. Try to include a joke")

# Get the follow-up response with full context
final_answer = chat(messages)
print(final_answer)
```
