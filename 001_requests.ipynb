```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

# Helper functions

These functions help us to store state in multi-turn conversations, allowing you to persist and send conversation history.

This is critical, since the Anthropic API is entirely stateless.

```python
def add_user_message(messages, text):
    messages.append({"role": "user", "content": text})

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    response = message.content[0].text
    messages.append({"role": "assistant", "content": response})
    return response
```

# Putting it all together

We can use these functions to begin a multi-turn conversation with the Anthropic API that maintains state.

```python
messages = []

add_user_message(messages, "Define quantum computing in one sentence")
chat(messages)

add_user_message(messages, "Write another sentence. Try to include a joke")
final_answer = chat(messages)
print(final_answer)
```
